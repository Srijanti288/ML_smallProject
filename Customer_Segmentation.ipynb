```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load dataset
# url = 'https://www.kaggle.com/datasets/kaushiksuresh147/customer-segmentation'
data = pd.read_excel('Online Retail.xlsx')

# Display first few rows of the dataset
data.head()
```

```python
# Checking for missing values
print(data.isnull().sum())

# Drop rows with missing CustomerID
data = data.dropna(subset=['CustomerID'])

# Drop irrelevant columns for segmentation
data = data[['CustomerID', 'Quantity', 'UnitPrice', 'Country']]

# Create a new column 'TotalPrice' as a product of Quantity and UnitPrice
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']

# Group data by CustomerID to get customer-level metrics
customer_data = data.groupby('CustomerID').agg({
    'Quantity': 'sum',
    'TotalPrice': 'sum',
    'Country': 'first'  # Retain the first country as Customer's location
}).reset_index()

# Display the cleaned and aggregated customer data
customer_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the numerical columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[['Quantity', 'TotalPrice']])

# Convert scaled data back into a DataFrame
scaled_customer_data = pd.DataFrame(scaled_data, columns=['Quantity', 'TotalPrice'])
scaled_customer_data.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow method to find optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(scaled_customer_data)
    wcss.append(kmeans.inertia_)

# Plot the elbow graph
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Choose optimal number of clusters (e.g., 4)
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(scaled_customer_data)

# Display customers with their cluster labels
customer_data.head()
```

```python
# Group by 'Cluster' and calculate the mean of only the numeric columns
numeric_columns = ['Quantity', 'TotalPrice']

# Use 'first' for 'Country' to keep the first country entry per cluster
cluster_summary = customer_data.groupby('Cluster').agg({
    'Quantity': 'mean',
    'TotalPrice': 'mean',
    'Country': 'first'  # Or use 'mode' if you want the most common country per cluster
})

# Display the cluster summary
print(cluster_summary)

# Analyze cluster size
customer_data['Cluster'].value_counts()

# Example: Visualizing clusters
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='TotalPrice', hue='Cluster', data=customer_data, palette='rainbow')
plt.title('Customer Segments Based on Quantity and TotalPrice')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages

# Create a PDF to store the plots and summary
with PdfPages('customer_segmentation_report.pdf') as pdf:

    # Scatter plot for Quantity vs TotalPrice colored by clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Quantity', y='TotalPrice', hue='Cluster', data=customer_data, palette='rainbow')
    plt.title('Customer Segments Based on Quantity and TotalPrice')
    pdf.savefig()  # Save the figure in the PDF
    plt.close()

    # Create bar chart for cluster sizes
    plt.figure(figsize=(10, 6))
    customer_data['Cluster'].value_counts().plot(kind='bar', color='skyblue')
    plt.title('Number of Customers in Each Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Number of Customers')
    pdf.savefig()  # Save the figure in the PDF
    plt.close()

    # Add cluster summary as a table in the PDF
    from pandas.plotting import table
    fig, ax = plt.subplots(figsize=(10, 4))  # Set the figure size
    ax.axis('off')  # Remove the axis
    tbl = table(ax, cluster_summary, loc='center', colWidths=[0.2]*len(cluster_summary.columns))  # Create a table
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(10)
    pdf.savefig()  # Save the table in the PDF
    plt.close()

print("PDF report generated: 'customer_segmentation_report.pdf'")
```
